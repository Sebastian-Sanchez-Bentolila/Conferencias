```python
# Convertir la columna Numero_de_ciclistas a numérica para análisis y gráficos
df["Numero_de_ciclistas"] = pd.to_numeric(df["Numero_de_ciclistas"], errors='coerce').fillna(0).astype(int)

# Generar gráficos
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Gráfico de barras para la cantidad de ciclistas por motivo
print(df.groupby("Motivo")["Numero_de_ciclistas"].sum())
df.groupby("Motivo")["Numero_de_ciclistas"].sum().plot(kind="bar", ax=axs[0], color="skyblue", edgecolor="black")
axs[0].set_title("Cantidad de Ciclistas por Motivo")
axs[0].set_xlabel("Motivo")
axs[0].set_ylabel("Cantidad de Ciclistas")

# Gráfico circular de anotaciones sobre el uso de la bicisenda
uso_bicisenda = df["Anotacion_1"].apply(lambda x: "No usa bicisenda" if "No usa la bicisenda" in x else "Usa bicisenda")
uso_bicisenda.value_counts().plot(kind="pie", autopct='%1.1f%%', startangle=140, ax=axs[1], colors=["lightcoral", "lightgreen"], wedgeprops={'edgecolor': 'black'})
axs[1].set_title("Distribución de Ciclistas según Uso de la Bicisenda")

plt.tight_layout()
plt.show()
```

```python
# Gráfico circular sobre uso de la bicisenda
fig, ax = plt.subplots(figsize=(6, 5))
uso_bicisenda = df["Anotacion_1"].apply(lambda x: "No usa bicisenda" if "No usa la bicisenda" in x else "Usa bicisenda")
uso_bicisenda.value_counts().plot(kind="pie", autopct='%1.1f%%', startangle=140, ax=ax, colors=["lightcoral", "lightgreen"], wedgeprops={'edgecolor': 'black'})
ax.set_title("Distribución de Ciclistas según Uso de la Bicisenda")

plt.tight_layout()
plt.show()
```

```python
total_ciclistas = df["Numero_de_ciclistas"].sum()
print("Total de ciclistas observados:", total_ciclistas)
```

```python
df.groupby("Motivo")["Numero_de_ciclistas"].sum().plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Cantidad de Ciclistas por Motivo")
plt.xlabel("Motivo")
plt.ylabel("Cantidad de Ciclistas")
plt.show()
```

```python
df["Anotacion_1"].value_counts().head(10).plot(kind="barh", color="lightseagreen", edgecolor="black")
plt.title("Comportamientos Observados de los Ciclistas")
plt.xlabel("Frecuencia")
plt.ylabel("Comportamiento")
plt.show()
```

```python
intervalos = df["Numero_de_ciclistas"].cumsum()
intervalos.plot(kind="line", color="purple")
plt.title("Acumulación de Ciclistas Observados cada 5 minutos")
plt.xlabel("Intervalos de 5 minutos")
plt.ylabel("Ciclistas acumulados")
plt.show()
```

```python
# Filtrar anotaciones que mencionan escuchar música
con_musica = df[df["Anotacion_1"].str.contains("música", case=False)]
sin_musica = df[df["Anotacion_1"].str.contains("música", case=False)]

# Conteo de ciclistas con y sin música por motivo
con_musica_count = con_musica.groupby("Motivo")["Numero_de_ciclistas"].sum()
sin_musica_count = sin_musica.groupby("Motivo")["Numero_de_ciclistas"].sum()

# Crear gráfico de barras apiladas
plt.figure(figsize=(10, 5))
plt.bar(con_musica_count.index, con_musica_count.values, label="Escuchando música", color="lightgreen")
plt.bar(sin_musica_count.index, sin_musica_count.values, label="Sin música", color="lightcoral", bottom=con_musica_count.values)
plt.title("Uso de Dispositivos de Entretenimiento por Motivo")
plt.xlabel("Motivo")
plt.ylabel("Cantidad de Ciclistas")
plt.legend()
plt.show()
```

```python
import seaborn as sns

# Crear tabla cruzada de motivos y comportamientos
comportamiento_por_motivo = pd.crosstab(df["Motivo"], df["Anotacion_1"])

# Crear el heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(comportamiento_por_motivo, cmap="YlGnBu", annot=True, fmt="d")
plt.title("Heatmap de Comportamiento por Motivo")
plt.xlabel("Comportamiento")
plt.ylabel("Motivo")
plt.show()
```
